```python
# 1.
# X = Income 
# Y = Life Expectancy 
# Bubble size = Population
# Color = World Region
# Year = 2010
```

```python

# 2.
df1 = pd.read_csv("Data/q1data.csv")
df1 = df1.sort_values(by='population', ascending=False)
df1 = df1[df1['year'] == 2010]
```

```python

region_colors = {
    'africa': '#04e2d7',   
    'americas': '#41e30a', 
    'asia': '#f54174',     
    'europe': '#f1e72d'
}

```

```python
# 3.
(ggplot(df1, 
        aes(x='income', 
            y='life_exp', 
            size='population', 
            fill='four_regions')) +
    geom_point() +
    
    scale_x_log10(
        # Adds the breaks 
        breaks=[1000, 2000, 4000, 8000, 16000, 32000, 64000], 
        limits = (500,64000),
        # Labels each break
        labels=['1000', '2000', '4000', '8000', '16k', '32k', '64k']) + 
    
    scale_y_continuous(
        # Adds limits to values between 15 and 90
        limits = (15,90),
        # Breaks y axis from 20 to 90 every 10
        breaks=range(20, 91, 10)) + 

    scale_size(range = (2,30)) +
    
    theme_minimal() + 
    labs(x='Income',
         y='Life expectancy',
         fill = 'World Regions') +
    
    scale_fill_manual(values=region_colors)+
    
    annotate('text', x=4000, y=40, label='2010', size=200, alpha=0.4, color='gray') +
    annotate('text', x=5000, y=90, label='Income Level 1        *        Level 2        *        Level 3            *        Level 4', size=20, alpha=0.5, color='Black') +
    annotate('text', x=500, y=90, label='years', size=10, alpha=0.4, color='black', angle=90) +

    guides(size=False) +

    theme(
        figure_size=(14,6)
    )
)
```

```python
# 4.
# Density plot using geom_density
(ggplot(df1, 
        aes(x='income', fill='four_regions')) +
geom_density(alpha=0.6) + 
scale_x_log10(
      breaks=[1000, 2000, 4000, 8000, 16000, 32000, 64000], 
      limits = (500,64000),
      # Labels each break
      labels=['1000', '2000', '4000', '8000', '16k', '32k', '64k']) +

scale_fill_manual(values=region_colors)+

scale_size(range = (2,30)) +
 theme_minimal() +
 labs(title='Density of Income by Region',
      x='Income',
      y='Density',
      color='Region'))

```

I tried using a density plot to display the same data but it does not visualize it as well. It cannot display two quantitative variables like a scatter plot, which makes it less effective at displaying the information because I am unable to see how Income and life expectency are related.

```python
# 1.
# X = Exports 
# Y = Imports 
# Bubble size = Energy Use
# Color = World Region
# Year = 1997
```

```python
# 2.
df2 = pd.read_csv("Data/q2data.csv")
df2['energy'] = df2['energy'].str.replace('k','').astype(float) * 1000
df2['energy'] = df2['energy'].astype(float)
df2 = df2.sort_values(by='energy', ascending=False)
df2 = df2.dropna()
```

```python
region_colors = {
    'africa': '#04e2d7',   
    'americas': '#41e30a', 
    'asia': '#f54174',     
    'europe': '#f1e72d'
}
```

```python
# 3.
(ggplot(df2, 
        aes(x='exports', 
            y='imports', 
            size='energy', 
            fill='four_regions')) +
    geom_point() +
    
    scale_x_continuous(
        limits = (0,220),
        breaks=range(0, 220, 20)) + 

    scale_y_continuous(
        limits = (0,400),
        breaks=range(0, 400, 50)) + 

    scale_size(range = (2,30)) +
    
    theme_minimal() + 
    labs(x='Export (% of GDP)',
         y='Imports (% of GDP)',
         fill = 'World Regions') +
    
    scale_fill_manual(values=region_colors)+
    
    annotate('text', x=90, y=200, label='1997', size=200, alpha=0.4, color='gray') +
    guides(size=False) +

    theme(
        figure_size=(14,6)
    )
)
```

```python
# 4. 
(ggplot(df2, aes(x='exports', y='imports', color='four_regions')) +
    geom_line(size=1) +  # Use line geometry
    theme_minimal() +
    scale_color_manual(values=region_colors) +
    labs(x='Exports (% of GDP)', 
         y='Imports (% of GDP)',
         color = 'World Regions') +     
    theme(figure_size=(14, 6))
)
```

`geom_line` is not a good geometric for displaying `exports` and `imports` because it implies an ordered relationship between points, which doesn't exist in this context. Since `exports` and `imports` are independent continuous variables without an sequential order, using a line can mislead by suggesting false connections or trends between them. A scatter plot or density plot is more appropriate for visualizing their relationship.

```python
# 1.
# X = Individuals using the Internet 
# Y = GDP/capita 
# Bubble size = Income
# Color = World Region
# Year = 2001
```

```python
# 2.
df3 = pd.read_csv("Data/q3data.csv")
df3 = df3.sort_values(by='income', ascending=False)
df3 = df3.dropna()
```

```python
# 3.
(ggplot(df3, 
        aes(x='internet_users', 
            y='gdp', 
            size='income', 
            fill='four_regions')) +
    geom_point() +
    
    scale_x_continuous(
        limits = (0,100),
        breaks=range(0, 90, 10)) + 
    
    scale_y_log10(
        # Adds the breaks 
        breaks=[200, 500, 1000, 2000, 5000, 10000, 20000, 50000, 100000], 
        limits = (200,150000),
        # Labels each break
        labels=['200', '500', '1000', '2000', '5000', '10k','20k', '50k', '100k']) + 

    scale_size(range = (2,30)) +
    
    theme_minimal() + 
    labs(x='Individuals using the Internet',
         y='GDP/capita',
         fill = 'World Regions') +
    
    scale_fill_manual(values=region_colors)+
    
    annotate('text', x=50, y=4000, label='2001', size=200, alpha=0.4, color='gray') +
    guides(size=False) +

    theme(
        figure_size=(14,6)
    )
)
```

```python
# 4.
(ggplot(df3, aes(x='internet_users', y='gdp')) +
    geom_density_2d(aes(color='four_regions'), size=1) + 
    scale_x_continuous(
        limits=(0, 100),
        breaks=range(0, 90, 10)) + 
    scale_y_log10(
        breaks=[200, 500, 1000, 2000, 5000, 10000, 20000, 50000, 100000], 
        limits=(200, 150000),
        labels=['200', '500', '1000', '2000', '5000', '10k', '20k', '50k', '100k']) + 
    scale_fill_manual(values=region_colors) +  
    scale_color_manual(values=region_colors) + 
    theme_minimal() + 
    labs(x='Individuals using the Internet (%)',
         y='GDP/capita',
         fill='World Regions',
         color='World Regions') +
    annotate('text', x=50, y=5000, label='2001', size=200, alpha=0.4, color='gray') + 
    guides(size=False) + 
    theme(figure_size=(14, 6))
)
```

A 2D density plot is good geometric for visualizing the relationship between `internet_users` and `GDP` because it displays areas where these variables have high or low concentrations. We can see where each World Regions peaks in `internet_users` and `GDP` and it helps identify patterns in their relationship. It is very similar to the scatter plot but removes individual points.
