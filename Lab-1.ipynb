```python

# 2
df1 = pd.read_csv("/Users/Steven/Documents/Documents - MacBook Pro/Cal Poly/MSBA/Fall 2024/GSB-544/Labs/Lab_1/Data/q1data.csv")
df1 = df1[df1['year'] == 2010]
df1 = df1.dropna()

# 3
(ggplot(df1, 
        aes(x='income', 
            y='life_exp', 
            size='population', 
            color='four_regions')) +
 geom_point(alpha=0.6) +
 scale_x_log10() + 
 scale_color_brewer(type='qual', palette='Set1') +  
 theme_minimal() +  # Use a minimal theme
 labs(title='Income vs. Life Expectancy by Population and Region',
      x='Income (log scale)',
      y='Life Expectancy',
      size='Population',
      color='Region'))


# 4
# Density plot using geom_density
(ggplot(df1, 
        aes(x='income', fill='four_regions')) +
 geom_density(alpha=0.6) +  # Use transparency for overlapping regions
 scale_x_log10() +  # Use log scale for better visualization
 theme_minimal() +
 labs(title='Density of Income by Region',
      x='Income (log scale)',
      y='Density',
      fill='Region'))
# I tried using a density plot to display the same data but it does not visualize it as well.
# It cannot display two quantitative variables like a scatter plot, which makes it less effective at displaying the information. 
```

```python

df2 = pd.read_csv("Data/q2data.csv")
df2 = df2[df2['year'] == 1997]
df2 = df2.dropna()

(ggplot(df2, 
aes(x = "exports", 
    y = "imports",
    size = "energy", 
    color = "four_regions"
))
+ geom_point()
)

# Import necessary libraries from plotnine
from plotnine import ggplot, aes, geom_point, scale_x_continuous, scale_y_continuous, theme_minimal, labs, scale_color_brewer

# Create the scatter plot using plotnine
(ggplot(df2, 
    aes(x = "exports", 
    y = "imports",
    size = "energy", 
    color = "four_regions")) +
 geom_point(alpha=0.6) +  # Use transparency for better visibility of overlapping points
 scale_x_continuous(name='Exports (% of GDP)') +  # Set x-axis label
 scale_y_continuous(name='Imports (% of GDP)') +  # Set y-axis label
 scale_color_brewer(type='qual', palette='Set1') +  # Set color palette
 theme_minimal() +  # Use a minimal theme
 guides(size=false) +
 labs(title='Imports vs. Exports as Percentage of GDP by Region',
      size='Energy Use',
      color='Region'))
```
